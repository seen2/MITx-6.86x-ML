```python
import numpy as np
import kmeans
import common
import naive_em
import em

X = np.loadtxt("toy_data.txt")

# TODO: Your code here
K=np.array([1,2,3,4])
seeds=np.array([0,1,2,3,4])
report={}
for k in K:
    report[k]=[]
    cost_prev=9**10
    for seed in seeds:
        while True:
            # print("k",k,"seed",seed)
            mu,var=common.init(X, k, seed)
            #Kmeans 
            # E
            post=kmeans.estep(X,mu)
            # M
            mixture,cost=kmeans.mstep(X,post)
            if cost<cost_prev:
                cost_prev=cost
            else:
                common.plot(X,mixture,post,f"K={k},seed={seed} cost={cost}")
                report[k].append(cost)
                break
            # common.plot(X,post,mixture,"Title")
       
print({key:min(report[key]) for key in report.keys()})
```

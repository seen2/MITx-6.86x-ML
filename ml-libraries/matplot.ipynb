```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x=np.arange(0,11)
y=np.arange(0,100)
y=0.5*x
```

```python
plt.plot(x,y,color="orange",linewidth=2,linestyle=":")
plt.plot(x,3*y,color="r",linewidth=4)
plt.title("X vs Y")
plt.xlabel("X-axis")
plt.ylabel("Y-axix")
plt.grid(True)
plt.show()
```

```python
#subplot
plt.subplot(1,2,1) #row,column,index
plt.plot(x,y,color="orange",linewidth=2,linestyle=":")
plt.title("X vs Y")
plt.xlabel("X-axis")
plt.ylabel("Y-axix")
plt.grid(True)
plt.subplot(1,2,2) #row,column,index
plt.plot(x,3*y,color="r",linewidth=4)
plt.title("X vs Y")
plt.xlabel("X-axis")
plt.ylabel("Y-axix")
plt.grid(True)
plt.show()
```

```python
#bar chart
student={'bob':70,'charlie':60,'ada':100}
names=student.keys()
marks=student.values()
plt.bar(names,marks)
plt.xlabel("Student Name")
plt.ylabel("Student Marks")
plt.grid(True)
plt.show()
```

```python
student={'bob':70,'charlie':60,'ada':100}
names=list(student.keys())
marks=list(student.values())
plt.barh(names,marks,color='r')
plt.ylabel("Student Name")
plt.xlabel("Student Marks")
# plt.grid(True)
plt.show()
```

```python
a=np.arange(0,100).tolist()
b=np.random.randint(1,200,100).tolist()
c=np.random.randint(1,200,100).tolist()
plt.scatter(a,b,marker='*',c='r',s=30)
plt.scatter(a,c,marker='.',c='b')
plt.grid(True)
plt.show()
```

```python
a=np.random.randint(1,100,100).tolist()
plt.hist(a,color='r',bins=30)
plt.grid(True)
plt.show()
```

```python
a=[10,30,40,20]
label=['A','B','C','D']
plt.pie(a,labels=label,autopct='%0.1f%%',colors=['red','blue','yellow','orange'])
plt.show()
```
